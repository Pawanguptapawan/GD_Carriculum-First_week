```python
#### 56. Generate a generic 2D Gaussian-like array (★★☆)

def generate_gau_arr(shape,mu,sigma):
    rows,cols=shape
    X=np.linspace(-1,1,rows)
    Y=np.linspace(-1,1,cols)
    x,y=np.meshgrid(X,Y)
    d=np.sqrt(x*x+y*y)
    g= np.exp(-((d-mu)**2//(2*sigma**2)))
    return g

g=generate_gau_arr((10,10),0.0,1.0)
g
    
```

```python
#### 57. How to randomly place p elements in a 2D array? (★★☆)

def put_p_elements(arr,p,value):
    rows,cols=arr.shape
    total_cells = rows * cols
    if p > total_cells:
        raise ValueError("Number of elements to place (p) cannot exceed total cells.")

   
    arr = np.zeros((rows, cols), dtype=int)

    
    flat_indices = np.arange(total_cells)
    
    
    chosen_flat_indices = np.random.choice(flat_indices, p, replace=False)

    
    chosen_rows, chosen_cols = np.unravel_index(chosen_flat_indices, arr.shape)

   
    arr[chosen_rows, chosen_cols] = value
    return arr

arr=np.random.random((10,10))
arr=put_p_elements(arr,5,1)
arr
```

```python

#### 93. Consider two arrays A and B of shape (8,3) and (2,2). How to find rows of A that contain elements of each row of B regardless of the order of the elements in B? (★★★)


A = np.array([
    [1, 2, 3],
    [3, 4, 5],
    [2, 1, 4],
    [6, 7, 8],
    [1, 3, 5],
    [2, 4, 6],
    [3, 1, 2],
    [4, 5, 6]
])

B = np.array([
    [1, 2],
    [4, 5]
])

mask = np.any(
    np.all(np.isin(A[:, None, :], B), axis=2),
    axis=1
)

result = A[mask]
print(result)
```
